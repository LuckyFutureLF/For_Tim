```python
import re
# Falls der Text noch Sonderzeichen beinhaltet / weiter bereinigt werden soll, können Sie die folgende Funktion anwenden (von Simon Köhl zur Verfügung gestellt):

"""
Custom cleaning function to clean text that is given in the wanted segments, meaning that it is pre-split.
@params: text --> list of strings which is the text split into the desired segments
@returns: desired segments as list of strings (each segment one string in the list) such that:
              - punctuation is removed
              - all text is lower case
              - interjection divider is removed
              - special characters such as /{}()\\n are removed
"""

def clean(text):
  # Tokenisierung mit .split()
  cleaned_segments=[] # neue Liste: wir packen die Wörter der Rede herein, nachdem wir sie im Folgenden Schritt pre-processen
  for segment in text: # für jedes Wort in der Rede
    segments = segment.split()
    cleaned_segment = []
    for wort in segments:
      wort = re.sub("\(\{[0-9]*\}\)", "", wort)
      wort=wort.lower().strip('"?!-/{()}.;,').split("\\n") # pre-processing = alle groß --> klein Buschstaben, Sonderzeichen weg, "\\n\\n" auch weg
      if len(wort)>1: # Die Wörter werden inszwischen einzeln in Listen gespeichert, dann schauen wir dass wir alle einzeln wieder in die End-Liste "cleaned_text" packen können
        for w in wort:
          w = re.sub("\(\{[0-9]*\}\)", "", w)
          if w != "":
            w = w.strip('"?!-/{()}.;,') # einige könnten noch Sondernzeichen beinhalten, deswegen double check
            cleaned_segment.append(w)
      else:
        if wort[0] != "":
          cleaned_segment.append(wort[0])
    seg_string = " ".join(cleaned_segment)
    if len(seg_string) != 0:
      cleaned_segments.append(" ".join(cleaned_segment))
  return cleaned_segments


# Anwendung der vorherigen Funktion "clean"
cleaned_text=clean(cleaned_text)
print("Die tokenisierte und pre-processed Rede: ",cleaned_text) # hier ist die pre-processed Version der Rede gespeichert, mit der Sie arbeiten können)
```
